```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
# Downloading Kompas data
headers = {'Accept-Language': 'en-US,en;q=0.8'}
url = 'https://www.cnbcindonesia.com/'
response = requests.get(url,headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
entries = soup.select('li article a')
#print(entries)
newstitle = [i.find('h2').get_text().strip() for i in entries]
print(newstitle)
print(len(newstitle))
print(" ")

urls = [i.attrs.get('href') for i in entries]
print(urls)
print(len(urls))
print(" ")

dates = []
times = []
categories = []
authors = []
#Look through each url for date, time, and author
for i in urls:
    text = BeautifulSoup(requests.get(i,headers=headers).text, "html.parser").find('div', class_ = 'date')
    if text is not None:
        text = text.get_text().strip().split()
        dates.append(' '.join(text[:-1]))
        times.append(text[-1])
    else:
        dates.append('-')
        times.append('-')
    text = BeautifulSoup(requests.get(i,headers=headers).text, "html.parser").find('div', class_ = 'author')
    if text is not None:
        texts = text.get_text().split('-')
        authors.append(texts[1].strip())
        categories.append(texts[0].strip())
    else:
        authors.append('-')
        categories.append('-')
        

print(authors)
print(times)    
print(dates)

df = pd.DataFrame(
    {'Title': newstitle,
     'Category': categories,
     'Author': authors,
     'Time': times,
     'Date': dates,
     'Link': urls}
    )

print(df.head())

df.to_csv('cnbcarticle.csv', index=False)
```

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
# Downloading Kompas data
headers = {'Accept-Language': 'en-US,en;q=0.8'}
url = 'http://books.toscrape.com/'
response = requests.get(url,headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
entries = soup.findAll('article', class_ = 'product_pod')
#print(entries)
titles = [i.find('h3').find('a').attrs.get('title') for i in entries]
print(titles)
print(len(titles))
print(" ")

prices = [i.find('p', class_ = 'price_color').get_text()[1:] for i in entries]
print(prices)
print(len(prices))
print(" ")

rDict = {'One' : 1,
        'Two' : 2,
        'Three' : 3,
        'Four' : 4,
        'Five' : 5}
ratings = [rDict[i.find(class_ = 'star-rating').attrs.get('class')[1]] for i in entries]
print(ratings)
print(len(ratings))
print(" ")

urls = ['http://books.toscrape.com/' +i.find('h3').find('a').attrs.get('href') for i in entries]
print(urls)
print(len(urls))
print(" ")

stocks = []
for i in urls:
    text = BeautifulSoup(requests.get(i,headers=headers).text, "html.parser").find(class_ = 'instock').get_text()
    stocks.append(text.split()[-2].strip('('))
print(stocks)

df = pd.DataFrame(
    {'Title': titles,
     'Price': prices,
     'Rating': ratings,
     'Stock': stocks,
     'Link': urls}
    )

print(df.head())

df.to_csv('books.csv', index=False)
```
